```python
""" Please edit this EV_dict to add the number of cars and their specifications as desired to give input to the code """
EV_dict = {
    'C1' : {
        'source' : 'V1',
        'destination': 'V8',
        'init_battery' : 5,
        'chrg_rate' : 4,
        'dischrg_rate' : 3,
        'max_battery_capacity' : 100,
        'avg_speed' : 10
    },
    "C2": {
        "source": "V2",
        "destination": "V7",
        "init_battery": 30,
        "chrg_rate": 10,
        "dischrg_rate": 15,
        "max_battery_capacity": 200,
        "avg_speed": 20
    },
    "C3": {
        "source": "V4",
        "destination": "V8",
        "init_battery": 4,
        "chrg_rate": 4,
        "dischrg_rate": 2,
        "max_battery_capacity": 100,
        "avg_speed": 15
    },
    "C4": {
        "source": "V5",
        "destination": "V1",
        "init_battery": 6,
        "chrg_rate": 2,
        "dischrg_rate": 2,
        "max_battery_capacity": 110,
        "avg_speed": 15
 
    }
}
""" Please edit city_edge_costs dictionary to add or remove cities as well as to specify the edges and their costs. """
city_edge_costs = {
    'V1' : {
        'V2' : 8,
        'V4' : 4,
        'V5' : 10
    },
    "V2": {
        "V1": 8,
        "V3": 4
    },
    "V3": {
        "V2": 4,
        "V4": 7,
        "V7": 6
    },
    "V4": {
        "V1": 4,
        "V3": 7,
        "V5": 8,
        "V7": 6
    },
    "V5": {
        "V1": 10,
        "V4": 7,
        "V6": 5
    },
    "V6": {
        "V5": 5,
        "V7": 1,
        "V8": 2
    },
    "V7": {
        "V3": 1,
        "V4": 4,
        "V6": 3,
        "V8": 5
    },
    "V8": {
        "V6": 2,
        "V7": 5
    }
}
 
cars = EV_dict.keys()
cities = city_edge_costs.keys()
BIG=1000000 #Initial cost value to be used in Dijkstra
```

```python
path_list=[]
time_list=[]
for car in cars:
    EV = EV_dict[car]
    maxcharge_distance = EV['max_battery_capacity']*EV['dischrg_rate']
    
    """ Modified Dijkstra: Finding the shortest distance and recording the shortest paths as we go to find the optimal 
    path later on in the code. """
    
    shortest_distance={}
    parent_nodes={}
    path=[]
    for city in cities:
        shortest_distance[city]=BIG
        if(city!=EV["source"]):
            path.append(city)
    shortest_distance[EV["source"]]=0
    parent_nodes[EV["source"]]=None
    
    n_cities = city_edge_costs[EV["source"]].keys()
    for n_city in n_cities:
        parent_nodes[n_city]=EV["source"]
        shortest_distance[n_city]=city_edge_costs[EV["source"]][n_city]
    while (len(path)):
        min_distance=BIG
        closest_city=None
        for city in path:
            dist= shortest_distance[city]
            if(dist<min_distance):
                min_distance=dist
                closest_city=city
        path.remove(closest_city)
        n_cities=city_edge_costs[closest_city].keys()
        for n_city in n_cities:
            if (city_edge_costs[closest_city][n_city]>min_distance):
                pass 
            upd=min_distance+city_edge_costs[closest_city][n_city]
            if (upd < shortest_distance[n_city]):
                shortest_distance[n_city]=upd
                parent_nodes[n_city]=closest_city
    optimal_path = []            
    path_link=EV['destination']
    
    """" Joining the optimal path links with the parent nodes to find the optimal path"""
    while(path_link!= None):        
                optimal_path.append(path_link)
                path_link = parent_nodes[path_link]
                optimal_time=0
                optimal_time=optimal_time+ (shortest_distance[EV['destination']]/EV['avg_speed']) 
                required_charge= (shortest_distance[EV['destination']]/EV['dischrg_rate']) - EV['init_battery']
                if(required_charge>0):
                    optimal_time = optimal_time + required_charge/EV['chrg_rate']

    optimal_path  = optimal_path[::-1]
    path_list.append(optimal_path)
    time_list.append(round(optimal_time,2))

    
car_list = list(EV_dict.keys())
for i in range(len(car_list)):   
    print("The optimal path for car {} is:".format(car_list[i]))
    print('\t',path_list[i])
    print('And the optimal journey time is: ',time_list[i],'\n')
    

    
```
